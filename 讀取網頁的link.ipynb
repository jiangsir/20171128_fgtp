```python
### 用 select 來過濾所要使用的 tag

import requests
from bs4 import BeautifulSoup
res = requests.get("http://apps.nknush.kh.edu.tw/ZeroBB")
soup = BeautifulSoup(res.text,'lxml')
for link in soup.select('a'):
    print(link.attrs)
    print(link.attrs['href'] +" : "+link.text)

    
for link in soup.find_all('a'):
    print(link)
```

```python
### 用 find_all 來過濾所要使用的 tag

import requests
from bs4 import BeautifulSoup
res = requests.get("http://apps.nknush.kh.edu.tw/ZeroBB")
soup = BeautifulSoup(res.text,'lxml')
    
for link in soup.find_all('a'):
    print(link)
```
